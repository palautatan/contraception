# Inference

## Possible roads to inference
- Chi-squared test for independence
    - `wife_education` x `husband education`
    - `standard_living` x `wife_work`
- Continuous variables compared between groups with ANOVA F test
    - `num_child` (this is `fertility_rate`) x `standard_living`
    - `kids_per_year` x `contraceptive`
    - Scramble the above

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colours_cafe = np.array(['#6B3231', '#DB565D', '#FACCAD', '#FF8A40'])
colours_cafe = sns.set_palette(sns.color_palette(colours_cafe))
```

```python
contra = pd.read_csv('../../1-data/train.csv')
```

```python
contra.shape
```

```python
contra.head()
```

# Fertility Rate

```python
contra.kids_per_year.apply(['min', 'mean', 'median', 'max'])
```

## Contraceptive Method

```python
contra['contraceptive'].value_counts()
```

```python
fig, ax = plt.subplots(1,2, sharey=True, figsize=(15,10))

sns.boxplot(x=contra['contraceptive'].map({0:'None', 1:'Short-term', 2:'Long-term'}),
            y='kids_per_year',
            data=contra,
            ax = ax[0])

ax[0].set(ylabel='Kids Per Year',
       xlabel='Contraceptive Method',
       title='Distribution of KPY per Contraceptive Method');

sns.violinplot(x=contra['contraceptive'].map({0:'None', 1:'Short-term', 2:'Long-term'}),
               y='kids_per_year',
               data=contra,
               ax = ax[1])

ax[1].set(ylabel='',
       xlabel='Contraceptive Method',
       title='Distribution of KPY per Contraceptive Method')


plt.show()
```

According to an ANOVA F test, there is a signficant difference in mean KPY between these groups.

```python
import scipy.stats as stats
anova = stats.f_oneway(contra['kids_per_year'][contra['contraceptive'] == 0], 
                       contra['kids_per_year'][contra['contraceptive'] == 1],
                       contra['kids_per_year'][contra['contraceptive'] == 2])
anova
```

```python
fig, ax = plt.subplots(1,2, sharey=True, figsize=(15,10))

sns.boxplot(x=contra['contraceptive'].map({0:'None', 1:'Short-term', 2:'Long-term'}),
            y='num_child',
            data=contra,
            ax = ax[0])

ax[0].set(ylabel='Fertility Rate',
       xlabel='Contraceptive Method',
       title='Distribution of Fertility Rate per Contraceptive Method');

sns.violinplot(x=contra['contraceptive'].map({0:'None', 1:'Short-term', 2:'Long-term'}),
               y='num_child',
               data=contra,
               ax = ax[1])

ax[1].set(ylabel='',
       xlabel='Contraceptive Method',
       title='Distribution of Fertility Rate per Contraceptive Method')


plt.show()
```

According to an ANOVA F test, there is a signficant difference in mean fertility rates between these groups.

```python
import scipy.stats as stats
anova = stats.f_oneway(contra['num_child'][contra['contraceptive'] == 0], 
                       contra['num_child'][contra['contraceptive'] == 1],
                       contra['num_child'][contra['contraceptive'] == 2])
anova
```

## Standard of Living

```python
contra['standard_living'].value_counts()
```

```python
fig, ax = plt.subplots(1,2, sharey=True, figsize=(15,10))

sns.boxplot(x=contra['standard_living'].map({1:'Low', 2:'Medium-low', 3:'Medium-high', 4:'High'}),
            y='kids_per_year',
            data=contra,
            order=['Low', 'Medium-low', 'Medium-high', 'High'],
            ax = ax[0])

ax[0].set(ylabel='Kids Per Year',
          xlabel='Standard of Living',
          title='Distribution of KPY per Standard of Living');

sns.violinplot(x=contra['standard_living'].map({1:'Low', 2:'Medium-low', 3:'Medium-high', 4:'High'}),
               y='kids_per_year',
               data=contra,
               order=['Low', 'Medium-low', 'Medium-high', 'High'], ax = ax[1])

ax[1].set(ylabel='',
       xlabel='Standard of Living',
       title='Distribution of KPY per Standard of Living')


plt.show()
```

By an ANOVA test, there **is a significant difference** in these groups' KPY.

```python
import scipy.stats as stats
anova = stats.f_oneway(contra['kids_per_year'][contra['standard_living'] == 1], 
                       contra['kids_per_year'][contra['standard_living'] == 2],
                       contra['kids_per_year'][contra['standard_living'] == 3],
                       contra['kids_per_year'][contra['standard_living'] == 4])
anova
```

```python
fig, ax = plt.subplots(1,2, sharey=True, figsize=(15,10))

sns.boxplot(x=contra['standard_living'].map({1:'Low', 2:'Medium-low', 3:'Medium-high', 4:'High'}),
            y='num_child',
            data=contra,
            order=['Low', 'Medium-low', 'Medium-high', 'High'],
            ax = ax[0])

ax[0].set(ylabel='Fertility Rate',
          xlabel='Standard of Living',
          title='Distribution of Fertility Rate per Standard of Living');

sns.violinplot(x=contra['standard_living'].map({1:'Low', 2:'Medium-low', 3:'Medium-high', 4:'High'}),
               y='num_child',
               data=contra,
               order=['Low', 'Medium-low', 'Medium-high', 'High'], ax = ax[1])

ax[1].set(ylabel='',
       xlabel='Standard of Living',
       title='Distribution of Fertility Rate per Standard of Living')


plt.show()
```

By an ANOVA test, there is no significant difference in these groups' fertility rates.

```python
import scipy.stats as stats
anova = stats.f_oneway(contra['num_child'][contra['standard_living'] == 1], 
                       contra['num_child'][contra['standard_living'] == 2],
                       contra['num_child'][contra['standard_living'] == 3],
                       contra['num_child'][contra['standard_living'] == 4])
anova
```
