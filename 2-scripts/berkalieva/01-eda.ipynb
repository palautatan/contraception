```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
sns.set(style="white", context="talk")
```

```python
# read in data
contra = pd.read_csv('../../1-data/contraceptive_for_students.csv')
```

```python
# refactor contraceptive use
this_dic = {1:0, 3:1, 2:2}
contra['contraceptive'] = contra['contraceptive'].map(this_dic)
```

```python
contra.head()
```

### Data Dictionary

The following is the data dictionary provided by the UCI Machine Learning Repository.

1. Wife's age (numerical)
2. Wife's education (categorical) 1=low, 2, 3, 4=high
3. Husband's education (categorical) 1=low, 2, 3, 4=high
4. Number of children ever born (numerical)
5. Wife's religion (binary) 0=Non-Islam, 1=Islam
6. Wife's now working? (binary) 0=Yes, 1=No
7. Husband's occupation (categorical) 1, 2, 3, 4
8. Standard-of-living index (categorical) 1=low, 2, 3, 4=high
9. Media exposure (binary) 0=Good, 1=Not good
10. Contraceptive method used (class attribute) 1=No-use, 2=Long-term, 3=Short-term

### Distribution of Contraceptive Use

```python
contra.groupby('contraceptive').contraceptive.count()
```

```python
plt.figure(figsize=(8,7))
ax = sns.barplot(x=contra.groupby('contraceptive').contraceptive.count().index,
                 y=contra.groupby('contraceptive').contraceptive.count() / contra.shape[0],
                 palette=['#27706B', '#DF6589', '#FFC3AF'])
ax.set(ylabel='Frequencies',
       xlabel='Contraceptive Method',
       title='Distribution of Contraceptive Method Choices');
```

### Overall Distribution of Women's Age

```python
plt.figure(figsize=(9,7))

ax = sns.distplot(contra['wife_age'], hist=False, color='#27706B')

ax.set(ylabel='Frequency',
       xlabel='Wife Age',
       title='Distribution of Age');
```

### Overall Distribution of Women's Standard of Living

```python
contra_standard = contra.groupby('standard_living').size().reset_index()
contra_standard = contra_standard.rename(columns={0:'count'})
```

```python
plt.figure(figsize=(9,7))

ax = sns.barplot(x = contra_standard['standard_living'], y = contra_standard['count'], color='pink')

ax.set(ylabel='Count',
       xlabel='Standard of Living',
       title='Breakdown of Standard of Living');
```

### Overall Distribution of Women's Education

```python
contra_ed = contra.groupby('wife_education').size().reset_index()
contra_ed = contra_ed.rename(columns={0:'count'})
```

```python
plt.figure(figsize=(9,7))

ax = sns.barplot(x = contra_ed['wife_education'], y = contra_ed['count'], color='pink')

ax.set(ylabel='Count',
       xlabel='Standard of Living',
       title='Breakdown of Woman Education');
```

### Standard of Living vs. Age

```python
plt.figure(figsize=(9,7))

ax = sns.distplot(contra[contra['standard_living']==1]['wife_age'], hist=False, color='#27706B', label='low')
ax = sns.distplot(contra[contra['standard_living']==2]['wife_age'], hist=False, color='#DF6589', label='medium-low')
ax = sns.distplot(contra[contra['standard_living']==3]['wife_age'], hist=False, color='#FFC3AF', label='medium-high')
ax = sns.distplot(contra[contra['standard_living']==4]['wife_age'], hist=False, color='black', label='high')

ax.set(ylabel='Frequency',
       xlabel='Wife Age',
       title='Distribution of Age by Standard of Living');
```

### Standard of Living vs. Religion

```python
contra_religion=contra.groupby(['wife_religion', 'standard_living']).size().reset_index()
contra_religion=contra_religion.rename(columns={0: "count"})
```

```python
plt.figure(figsize=(9,7))

ax = sns.barplot(x='wife_religion', y='count', hue='standard_living', data=contra_religion,
                 palette=['#27706B', '#DF6589', '#FFC3AF', 'pink'])

ax.set(ylabel='Count',
       xlabel='Wife Religion',
       title='Breakdown of Islam Religion by Standard of Living');
```

```python
plt.figure(figsize=(9,7))

ax = sns.barplot(x='standard_living', y='count', hue='wife_religion', data=contra_religion,
                 palette=['#DF6589', '#FFC3AF'])

ax.set(ylabel='Count',
       xlabel='Wife Religion',
       title='Breakdown of Standard of Living by Islam Religion');
```

### Standard of Living vs. Education

```python
contra_ed = contra.groupby(['wife_education', 'standard_living']).size().reset_index()
contra_ed = contra_ed.rename(columns={0: "count"})
```

```python
plt.figure(figsize=(9,7))

ax = sns.barplot(x='wife_education', y='count', hue='standard_living', data=contra_ed,
                 palette=['#27706B', '#DF6589', '#FFC3AF', 'pink'])

ax.set(ylabel='Count',
       xlabel='Wife Education',
       title='Breakdown of Wife Education by Standard of Living');
```

```python
plt.figure(figsize=(9,7))

ax = sns.barplot(x='standard_living', y='count', hue='wife_education', data=contra_ed,
                 palette=['#27706B', '#DF6589', '#FFC3AF', 'pink'])

ax.set(ylabel='Count',
       xlabel='Standard of Living',
       title='Breakdown of Standard of Living by Wife Education');
```
